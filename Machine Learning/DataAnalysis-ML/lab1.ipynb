```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_csv('lab1PL.csv', delimiter=';')
df = df.dropna()
mean_value = df['thsd.USD'].mean()
median_value = df['thsd.USD'].median()
mode_value = stats.mode(df['thsd.USD'], keepdims=True)[0][0]
dispersion_value = np.var(df['thsd.USD'])
std_dev_value = np.sqrt(dispersion_value)
```

```python
print(f"Математичне сподівання: {mean_value}")
print(f"Медіана: {median_value}")
print(f"Мода: {mode_value}")
print(f"Дисперсія: {dispersion_value}")
print(f"Середньоквадратичне відхилення: {std_dev_value}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['thsd.USD'], bins=10)
plt.xlabel('thsd. USD')
plt.ylabel('Frequency')
plt.title('Гістограма розподілу значень thsd. USD')
plt.show()
```

```python
# операції з Series
series = df['thsd.USD']
print("Перших 5 елементів:")
print(series.head())
print("Останніх 5 елементів:")
print(series.tail())
print("Розмірність:", series.shape)
print("Типи даних:", series.dtypes)
print("Кількість унікальних значень:", series.nunique())
print("Загальна статистики:")
print(series.describe())
```

```python
# Операції з DataFrame
print("Перших 5 рядків:")
print(df.loc[:4])  
print("Останніх 5 рядків:")
print(df.iloc[-5:]) 
print("Розмірність:", df.shape)
print("Типи даних:")
print(df.dtypes)
print("Загальна статистика:")
print(df.describe())
df_sorted_by_column = df.sort_values(by="thsd.USD", ascending=True)
print("Сортований DataFrame за 'thsd.USD' (зростання):")
print(df_sorted_by_column.head())
df_sorted_by_index = df.sort_index(ascending=False)
print("Сортований DataFrame за індексами у зворотному порядку:")
print(df_sorted_by_index.head())
print("Шостий регіон:", df.iat[5, 0])
print("Значення thsd.USD шостого регіону:", df.iat[5, 1])
print(df.T)
```

```python
titanic = pd.read_csv('https://vincentarelbundock.github.io/'+'Rdatasets/csv/carData/TitanicSurvival.csv')
```

```python
pd.set_option('display.precision', 2)
print(titanic.head())
print(titanic.tail())
```

```python
titanic.columns = ['name', 'survived', 'sex', 'age', 'class']
print(titanic.head())
youngest = titanic["age"].min()
oldest = titanic["age"].max()
mean_age = titanic["age"].mean()
survival_stats = titanic["survived"].value_counts()
```

```python
print(f"\nСамий молодий пасажир: {youngest} років")
print(f"Самий старий пасажир: {oldest} років")
print(f"Средній вік пасажирів: {mean_age:.2f} років")
print("Статистика по вижившим:", survival_stats)
```

```python
women_first_class = titanic[(titanic["sex"] == "female") & (titanic["class"] == "1st")]

youngest_woman = women_first_class["age"].min()
oldest_woman = women_first_class["age"].max()

survived_women_first_class = women_first_class["survived"].value_counts()
```

```python
print(f"\nСама молода жінка 1-го класса: {youngest_woman} років")
print(f"Сама старая жінка 1-го класса: {oldest_woman} років")
print(f"Кількість виживших жінок 1-го класса: {survived_women_first_class.get('yes', 0)}")
```

```python
titanic.hist(column="age", bins=20, edgecolor="black", figsize=(8, 6))
plt.xlabel("Вік")
plt.ylabel("Кількість пасажирів")
plt.title("Гістогрма віку пасажирів")
plt.grid(False)
plt.show()
```
